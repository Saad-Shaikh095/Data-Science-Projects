```python
import pandas as pd
```

```python
df=pd.read_csv("Bmi_male_female.csv")
```

```python
df.head()
```

```python
x=df[["Height","Weight"]]
y=df.Gender
```

```python
Y=pd.get_dummies(y)
```

```python
Y
```

```python
x
```

```python
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x,Y)
```

```python
ans=model.predict([[130,90]])
```

```python
if ans[0][0]==True:
    print("Female")
else:
    print("Male")
```
